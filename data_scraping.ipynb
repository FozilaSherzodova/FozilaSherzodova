```python
import requests
from bs4 import BeautifulSoup
```

```python
#1 BOOKS TO SCRAP
url = "https://books.toscrape.com"
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, "html.parser")
titles = soup.find_all("h3")
for title in titles:
    print(title.text)
```

```python
base_url = "http://books.toscrape.com/catalogue/page-{}.html"
for page in range(1, 51): 
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    books = soup.find_all("article", class_="product_pod")
 
for book in books:
       title = book.h3.a["title"]
        
       price = book.find("p", class_="price_color").text[1:]
        
       rating_class = book.find("p", class_="star-rating")["class"]
       rating = rating_class[1] 
       print(f"title: {title}\nrating: {rating}\nprice: {price}\n")
```

```python
#2 QUOTES TO SCRAPE
quote_url = "https://quotes.toscrape.com/"
r = requests.get(quote_url)
soup = BeautifulSoup(r.text, "html.parser")
quotes = soup.find_all(class_='quote')
quotes
for quote in quotes:
    text = quote.find(class_='text').get_text()
    author = quote.find(class_='author').get_text()
    tags = [tag.get_text() for tag in quote.find_all(class_='tag')]
    print('--------------------')
    print("The Quote",text) 
    print("Author: ", author)
    print("Tags: ", ', '.join(tags))
   
```

```python
#5 Coin Market Cap
cap_url = 'https://coinmarketcap.com/'
cap_response = requests.get(cap_url)
cap_soup = BeautifulSoup(cap_response.text, "html.parser")
# cap_soup.find_all(class_ = "sc-65e7f566-0 iPbTJf coin-item-name")

rows = cap_soup.find_all('tr')

table = []

for row in rows:
    cells = row.find_all(['th', 'td'])
    table.append([cell.text for cell in cells])


for i in range(1,20):
    print("\n-------------------\n")
    print(f"{i}Name: {table[i][2]}\nPrice: {table[i][3]}\n1h%: {table[i][4]}")


```

```python
import requests
from bs4 import BeautifulSoup

w_url = "https://weather.com/uz-UZ/ob-havo/bugun/l/UZXX0004:1:UZ?Goto=Redirected"

w_response = requests.get(w_url, "html.paeser")
w_soup = BeautifulSoup(w_response.text, "html.parser")

temperature = w_soup.find(class_="TodayDetailsCard--feelsLikeTempValue--8WgHV")
details = w_soup.find_all(class_="WeatherDetailsListItem--wxData--lW-7H")

if temperature and len(details) >= 2:
    namlik = details[2].text
    pressure = details[4].text

    print(f"Now the temperature is {temperature.text}")
    print(f"Now the namlik (humidity) is {namlik}")
    print(f"Now the pressure is {pressure}")
else:
    print("Weather details not found or the structure of the page has changed.")
```
